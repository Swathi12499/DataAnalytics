# Data Visualisation

**Histogram**

```
import matplotlib.pyplot as plt
fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['Rings'],bins =6) # number of bins
plt.title('Abalone distribution')
plt.xlabel('Rings')
plt.ylabel('#Abalones')
plt.show()
```

**Boxplots**

```
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
ax.boxplot(df["Rings"])
plt.show()
```

```
data.boxplot(column=["Length","Diameter","Shell weight","Whole weight"])
```

**Scatter plots**

```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['Shell weight'],df['Rings']) #You can also add more variables here to represent color and size.
plt.show()
```

```
plots.scatter(data.Diameter,data.Rings)
```

**Line Chart**

```
var = df.groupby('Sex').Rings.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Gender')
ax1.set_ylabel('Rings')
ax1.set_title("Gender- Ring distribution")
var.plot(kind='line')
```
